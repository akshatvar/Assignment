```python
#Explain why we have to use the Exception class while creating a Custom Exception

#Solution
"""
Custom exceptions provide you the flexibility to add attributes and methods that are not
part of a standard Python exception. These can store additional information, like an
application-specific error code, or provide utility methods that can be used to handle
or present the exception to a user.
"""
#let's take an example to understand this
"""
create a function which take phone number as an input, since phone number can't be less than
and greater than 10 digit
"""
class validPhoneNum(Exception):
    def __init__(self,num):
        self.num = num
```

```python
def checkPhoneNum(num):
    NumLen = len(num)
    if NumLen != 10:
        raise validPhoneNum("Phone Number is Invalid")
    else:
        logging.error("Phone Number is valid")
```

```python
try:
    Phone_num = input("Enter Phone number")
    checkPhoneNum(Phone_num)
    print("try is excecuted")
except Exception as e:
    logging.error(e)
```

```python
#What errors are defined in the ArithmeticError class? Explain any two with an example.

#Solution
"""
The errors defined in Arthemetic Error class are given below
OverflowError
ZeroDivisionError
FloatingPointError
DecimalException
"""
#ZeroDivisionError
"""
A ZeroDivisionError occurs in Python when a number is attempted to be divided by zero.
Since division by zero is not allowed in mathematics, attempting this in Python code raises
a ZeroDivisionError
"""
try:
    num = int(input("Enter Numerator"))
    den = int(input("Enter Denominator"))
    ans = num/den
except ArithmeticError as e:
    logging.error("This statement is raising an error".format(e))
else:
    logging.error(ans)
    logging.error("Success")
```

```python
#OverFlowError
"""
When an arithmetic operation exceeds the limits of the variable type, an OverflowError is
raised. Long integers allocate more space as values grow, so they end up raising MemoryError.
Floating point exception handling is not standardized, however. Regular integers are converted
to long values as needed.
"""
try:
    i = 1
    f = 3.0**i
    for i in range(1,100):
        print(i,f)
        f = f**2
except ArithmeticError as e:
    logging.error("Overflowed")
```

```python
#Why LookupError class is used? Explain with an example KeyError and IndexError

#Solution
"""
LookupError Exception is the Base class for errors raised when something canâ€™t be found.
The base class for the exceptions that are raised when a key or index used on a mapping
or sequence is invalid: IndexError, KeyError.
"""
#Index Error:- An IndexError is raised when a sequence reference is out of range.
try:
    l = ['a','b','c','d','e','f']
    n = int(input("Enter Index"))
    print(l[n])
except LookupError as e:
    logging.error(e)
```

```python
#Key Error:- A KeyError is raised when a value is not found as a key of a dictionary.
don = {'name':'Don Bhai','Address':'Chor Bazaar','Crime':'Bahut Saare'}
k = input("What do you want to know about Don")
try:
    print(f'{k} is {don[k]}')
except KeyError as e:
    logging.error("KeyErrorOccur occur doesn't found".format(e))
```
