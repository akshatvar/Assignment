```python
#Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

#Solution
"""
1.Bar Plot:
Use: Bar plots are used to visualize categorical data with rectangular bars. They can be used to compare the frequency
or value of different categories.
Example Use Cases: Showing the sales of different products, comparing the performance of different teams in a competition,
displaying the distribution of categorical variables.

2.Histogram:
Use: Histograms are used to represent the distribution of a continuous variable. They display the frequency of data points
within specified bins.
Example Use Cases: Analyzing the distribution of exam scores, understanding the age distribution of a population, studying
the distribution of income levels.

3.Scatter Plot:
Use: Scatter plots are used to visualize the relationship between two continuous variables. Each data point is represented as a
dot, and patterns or trends can be observed.
Example Use Cases: Exploring the correlation between height and weight, analyzing the relationship between study hours and exam scores,
identifying clusters in data points.

4.Box Plot:
Use: Box plots (box-and-whisker plots) display the distribution of a dataset's summary statistics, including median, quartiles,
and outliers. They are useful for identifying the spread and skewness of the data.
Example Use Cases: Comparing the distribution of salaries in different job roles, analyzing the spread of temperatures in different
seasons, understanding the variation in housing prices across neighborhoods.

5.Heatmap:
Use: Heatmaps are used to visualize the correlation between different variables in a dataset. They use color intensity to represent
the strength of the relationship between variables.
Example Use Cases: Analyzing the correlation between different features in a dataset, identifying which factors are most 
influential in a multi-dimensional dataset, exploring the co-occurrence of different items.
```

```python
import seaborn as sn
```

```python
import matplotlib.pyplot as plt
```

```python
# #Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
# "timepoint" and y = "signal" for different events and regions.

#Solution
fmri = sns.load_dataset('fmri')
sns.lineplot(x=fmri.timepoint,y=fmri.signal)
```

```python
#Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
#'pclass', y = 'age' and y = 'fare'.

#Solution
titanic = sns.load_dataset('titanic')

plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
sns.boxplot(x=titanic.pclass,y=titanic.age)
plt.title("plot age by pclass")

plt.subplot(1,2,2)
sns.boxplot(x=titanic.pclass,y=titanic.fare)
plt.title("plot age by fare")
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
# parameter for the 'cut' column of the diamonds dataset.

#Solution
diamond = sns.load_dataset('diamonds')
sns.histplot(data = diamond,x='price',hue='cut',multiple='stack',bins=30)
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
# of the iris dataset.

#Solution
iris = sns.load_dataset('iris')
sns.pairplot(data = iris,hue='species')
plt.suptitle("Pair Plot of Iris Dataset by Species", y=1.02)
plt.show()
```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

#Solution
flight = sns.load_dataset('flights')
flight_pivot = flight.pivot_table(index='month',columns='year',values='passengers')
plt.figure(figsize=(10, 7))
sns.heatmap(flight_pivot,cmap="#E6D421", annot=True, fmt="d")
plt.title("Passenger Count Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
