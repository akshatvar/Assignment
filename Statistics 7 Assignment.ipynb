#Question 1.

Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation 
of 5 using Python. Interpret the results.

#Solution

```python
import scipy.stats as stat

#Let assume sample size is 100
sample_size = 100
mean = 50
std_dev = 5
confidence_interval = 0.95
significance_level = 1 - confidence_interval

#Calculate Standard error
standard_error = std_dev/sample_size**0.5

#Calculate margin of error
margin_of_error = stat.norm.ppf(1-(1-significance_level)/2)*standard_error

#Confidence Interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Confidence Interval lie between {lower_bound:.2f} to {upper_bound:.2f}")
```

#Question 2

Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag 
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% 
brown. Use Python to perform the test with a significance level of 0.05

#Solution

```python
import scipy.stats as stat

#Null Hypothesis
H0 = 'The distribution of colors of M&Ms are matches the expected distrbution'
#Altenative Hypothesis
H1 = 'The distribution of colors of M&Ms are not matches the expected distrbution'

#let observed data be
observed_data = [25,30,15,10,15,5]
expected_data = [20,20,20,10,10,20]

chisquare_test, p_value = stat.chisquare(observed_data,expected_data)

significance_value = 0.05
dof = len(expected_data) - 1
critical_value = stat.chi2.ppf(1-significance_value,dof)

if critical_value < chisquare_test:
    print("Reject the Null Hypothesis: The distribution of colors of M&Ms are not matches the expected distrbution")
else:
    print("Fail to Null Hypothesis: The distribution of colors of M&Ms are matches the expected distrbution")
```

#Question 4.

A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals 
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the 
population who smoke.

#Solution

```python
import scipy.stats as stat
import math
sample_space = 500
x = 60

#proportion of people who smoke
p = x/sample_space

confidence_level = 0.95

standard_error = math.sqrt((p*(1-p))/sample_space)

margin_of_error = stat.norm.ppf((1 + confidence_level)/2)*standard_error

lower_bound = p - margin_of_error
upper_bound = p + margin_of_error

lower_bound_percentage = lower_bound*100
upper_bound_percentage = upper_bound*100

print(f"confidence interval for true proportion lie between {lower_bound_percentage:.2f}% to {upper_bound_percentage:.2f}%")
```

#Question 6.

Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the 
area corresponding to a chi-square statistic of 15

```python
import scipy.stats as stat
import numpy as np
import pandas as pd
import plotly.graph_objects as go

dof = 10

#create data set
x = np.linspace(0,30,2000)

#create probability density function
pdf = stat.chi2.pdf(x,dof)

# Define the chi-square statistic for shading
chi_square_statistic = 15

# Create a DataFrame for the plot
data = {'x': x, 'PDF': pdf}
df = pd.DataFrame(data)

fig = go.Figure()
fig.add_trace(go.Scatter(x=df['x'],y=df['PDF'], mode='lines', name=f'Chi-Square ({dof} degrees of freedom)'))
shaded_area = df[df['x'] >= chi_square_statistic]
fig.add_trace(go.Scatter(x=shaded_area['x'],y=shaded_area['PDF'],fill='tozeroy', fillcolor='lightblue', name=f'X >= {chi_square_statistic}'))
fig.show()
```

#Question 7.

A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 
preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who 
prefer Coke

#Solution

```python
import scipy.stats as stat
import math

sample_space = 1000
x = 520

#proportion of people who preffered coke
p = x/sample_space

confidence_level = 0.99

#Calculate standard error
standard_error = math.sqrt(p*(1-p))/sample_space

#Calculate margin of error
margin_of_error = stat.norm.ppf((1+confidence_level)/2)*standard_error

lower_bound = p - margin_of_error
upper_bound = p + margin_of_error

lower_bound_percentage = lower_bound*100
upper_bound_percentage = upper_bound*100

print(f"confidence interval for true proportion lie between {lower_bound_percentage:.2f}% to {upper_bound_percentage:.2f}%")
```

#Question 8.

A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 
45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the 
expected frequencies of a fair coin. Use a significance level of 0.05.

#Solution

```python
import scipy.stats as stats

#Below data represent no. heads and tails
observed_data = [55,45]
expected_data = [50,50]

H0 = 'The coin is fair'
H1 = 'The coin is baised toward the tail'

chisquare_test, p_value = stats.chisquare(observed_data,expected_data)
significance_value = 0.05
dof = len(observed_data) - 1

critical_value = stats.chi2.ppf(1 - significance_value,dof)

if chisquare_test > critical_value:
    print(f"Reject the Null Hypothesis: {H1}")
else:
    print(f"Fail to Reject the Null Hyothesis: {H0}")
```

#Question 9. 

A study was conducted to determine if there is an association between smoking status (smoker or 
non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below. 
Conduct a chi-square test for independence to determine if there is a significant association between 
smoking status and lung cancer diagnosis.
Use a significance level of 0.05

Lung Cancer    Yes     No
Smoker          60     140
Non-smoker      30     170

#Solution

```python
import scipy.stats as stats
import numpy as np

H0 = "No, There is not a significant association between smoking status and lung cancer diagnosis"
H1 = "Yes, There is a significant association between smoking status and lung cancer diagnosis"

contingency_table = np.array([[60,140],
                             [30,170]])

chisquare_test, p_value, dof, expected_frequency = stats.chi2_contingency(contingency_table)

significance_value = 0.05

critical_value = stats.chi2.ppf(1-significance_value,dof)

if chisquare_test > critical_value:
    print(f"Reject the Null Hypothesis: {H1}")
else:
    print(f"Fail to Reject the Null Hyothesis: {H0}")
```

Question 11. 

A random sample of 30 people was selected from a population with an unknown mean and standard 
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. 
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a 
significance level of 0.05

#Solution

```python
import scipy.stats as stats

sample_size = 30
sample_mean = 72
sample_stddev = 10
population_mean = 70
significance_level = 0.05

H0 = "The population mean (μ) is equal to 70."
H1 = "The population mean (μ) is not equal to 70."

t_statistic = (sample_mean - population_mean)/(sample_stddev/(sample_size**0.5))

dof = sample_size-1
alpha = significance_level/2 #for two tailed test

critical_value_lower = stats.t.ppf(alpha,dof)
critical_value_upper = stats.t.ppf(1-alpha,dof)

if t_statistic < critical_value_lower or t_statistic > critical_value_upper:
    print(f"Reject the Null Hypothesis: {H1}")
else:
    print(f"Fail to Reject the Null Hypothesis: {H0}")
```
