```python
#Q5. Explain the concept of levels of measurement and give an example of a variable for each level.
"""
Nominal
The first level of measurement is nominal. Nominal data is data that falls into categories, like ethnicity, religion,
and gender.

Here are examples of questions that measure nominal data:

What is your occupation?
Are you employed or unemployed?
If there are only two categories of answers, then it is called a dichotomy, like yes and no questions.

Here’s an example of a question that measures nominal data and is a dichotomy:
Did you miss a day of work this month? Yes or no?

Ordinal

The second level of measurement is ordinal. Ordinal data has an order or rank to it, and you can tell that one answer
is more than or less than another. For example, 1st place, 2nd place, and 3rd place in a race. You know that 1st place
is something more than the others. Likert scales, where you answer based on a scale of 1-5 where 1 is strongly disagree
and 5 is strongly agree is another example of ordinal data.

Here’s an example of a question that measures ordinal data.

What level of education do you have? no high school degree, high school degree, some college, college degree, postgraduate
education. (These categories can be ranked in order of education, so they are ordinal.)

Interval

The third level of measurement is interval. Interval data is numbers that represent positions along a continuous number line,
but there is something called no true zero. This means that interval data can be zero, but that zero does not really mean zero
as in nothing. For example, a thermometer can read zero, but that doesn’t mean there is zero temperature.

Interval data shows a relative position, so like a reading on the thermometer shows a position relative to the other places on the
thermometer. And the intervals between each section on the thermometer are the same—for example, the distance between 36 degrees
Celsius and 38 degrees Celsius is the same as the distance between 40 and 42 degrees.

Here’s an example of a question that measures interval data:

Please rate our company on a scale of 0 to 10.

Ratio

The fourth level of measurement is ratio. Ratio data is just like interval data, but it DOES have a true zero, which is a zero
point that really means “nothing.” Some examples of ratio data are income, distance traveled to work, and number of cigarettes
smoked per day. If you have zero income, that really means you have nothing. If you traveled zero miles to work, then you really 
didn’t go anywhere. If you put down zero cigarettes, then you really smoked no cigarettes. Zero with ratio data really means
zero—nothing.

An example of a question that involves ratio data is:

How many employees work for you? (and there is a blank spot to write the number.)
"""
```

```python
#Q9. Describe the difference between descriptive and inferential statistics. Give an example of each
#type of statistics and explain how they are used.
"""
Descriptive statistics is the branch of statistics that deals with the collection, organization, analysis, interpretation,
and presentation of data. It provides a summary of the data and describes the main features of the data set, such as the mean,
median, mode, and standard deviation.

Suppose you have collected the ages of 50 individuals in a study group. You can calculate the mean (average) age, the median
(middle value), the range (difference between the maximum and minimum ages), and the standard deviation (measure of how spread
out the ages are). These summary statistics provide a snapshot of the distribution of ages within your study group

Inferential statistics, on the other hand, is the branch of statistics that deals with
making predictions or inferences about a population based on a sample of data. It uses statistical methods to make estimates
and predictions about a population based on a sample of data, such as estimating the proportion of a population that holds a
certain viewpoint or the mean income of a population.In summary, descriptive statistics provides information about a sample
of data, whereas inferential statistics uses that information to make predictions about a population.

Continuing from the age example, let's say you want to determine if the average age of your study group is significantly different
from the average age of the general population. You would collect your sample's data and then perform hypothesis testing to determine
if any observed differences are likely due to chance or if they are statistically significant.
"""
```
