```python
#Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
#Matplotlib.

#Solution
"""
Matplotlib is a popular data visualization library in Python. It provides a wide variety of functions and tools for
creating static, interactive, and animated visualizations in Python. Matplotlib is widely used for generating plots,
charts, graphs, and other visual representations of data.

It is used for:

1.Data Visualization: Matplotlib allows you to create various types of plots and charts to visualize data, making
it easier to understand trends, patterns, and relationships within the data.

2.Presentation and Reporting: It is commonly used to create visualizations for presentations, reports, and publications,
enhancing the communication of findings and insights.

3.Exploratory Data Analysis (EDA): Matplotlib is useful for exploring data and gaining initial insights, which can guide
further analysis and decision-making.

4.Model Evaluation and Interpretation: It can be used to visualize model predictions, performance metrics, and decision
boundaries, aiding in model evaluation and interpretation.

5.Customization and Styling: Matplotlib provides a high level of customization for plot appearance, enabling users to
tailor visualizations to their specific needs.

Five plots that can be created using the Pyplot module of Matplotlib are:

1.Line Plot: A simple plot of data points connected by straight lines, often used to represent trends over time or
across a continuous variable.

2.Bar Plot: A plot where rectangular bars are used to represent data values. It's commonly used to compare the values
of different categories.

3.Histogram: A graphical representation of the distribution of a dataset, showing the frequency or count of data points within specified bins.

4.Scatter Plot: A plot where individual data points are represented as dots on a two-dimensional plane. It's used to visualize the
relationship between two variables.

5.Pie Chart: A circular chart divided into sectors, where each sector represents a proportion of a whole. It's suitable for
displaying relative proportions or percentages.
```

```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
#plot a scatter plot.

#Solution
"""
A plot where individual data points are represented as dots on a two-dimensional plane. It's used to visualize the
relationship between two variables.
"""
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("Value of x")
plt.ylabel("Value of y")
plt.title("Scatter Plot")
plt.show()
```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

#Solution
"""
The subplot() function in Matplotlib is used to create multiple plots within the same figure, allowing you to arrange
multiple plots in a grid layout. This is useful when you want to compare different visualizations side by side or display
multiple views of the same data.
"""
#Use the following data:
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title("z1")

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title("z2")

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title("z3")

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title("z4")

plt.tight_layout()
plt.show()
```

```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
#import numpy as np
# company = np.array(["Apple", "Microsoft", "Google", "AMD"])
# profit = np.array([3000, 8000, 1000, 10000])

#Solution
"""
A bar plot is a type of data visualization that uses rectangular bars to represent data values. The length or height
of each bar is proportional to the value it represents.
Bar plots are used to compare and visualize the distribution
or relationship of categorical data or discrete values. They are particularly useful for showing the relative sizes
of different categories or groups.
"""
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()
```

```python
#  Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# # box1 = np.random.normal(100, 10, 200)
# # box2 = np.random.normal(90, 20, 200)

#Solution
"""
A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of the
distribution of a dataset. It displays key statistical measures such as the median, quartiles, and potential outliers.
The box plot consists of a box that represents the interquartile range (IQR) and "whiskers" that extend from the box to
indicate the range of data within a certain factor of the IQR. Outliers, which are data points significantly different
from the rest of the data, are sometimes also shown individually as points or dots.
"""
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]
lables = ['Box1','Box2']
plt.figure(figsize=(6,6))
plt.boxplot(data,labels=lables)
plt.ylabel("Values")
plt.title('Box Plot for Box 1 and Box 2')
plt.show()
```
