```python
#Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects
#all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The
#function should return a new DataFrame that contains only the selected rows.

#Solution
data8 = {"A":[3,7,3,6,2,9,10,6,7,11]
        ,"B":[6,9,3,6,11,5,3,8,10,9]
        ,"C":[8,6,5,7,3,9,10,3,4,2]}
df8 = pd.DataFrame(data8)
```

```python
df8
```

```python
def filter(df):
    filter_df = df[(df['A']>5) & (df['B']<10)]
    return filter_df
filter_df = filter(df8)
```

```python
#Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,
#median, and standard deviation of the values in the 'Values' column.

#Solution
total_bill_mean = df['total_bill'].mean()
total_bill_median = df['total_bill'].median()
total_bill_std = df['total_bill'].std()

total_bill_mean = round(total_bill_mean, 6)
total_bill_median = round(total_bill_median, 6)
total_bill_std = round(total_bill_std, 6)

print("mean of total bill",total_bill_mean)
print("median of total bill",total_bill_median)
print("standard deviation of total bill",total_bill_std)
```

```python
#Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
#function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

#Solution
date_data1 = {'Date': ['2023-01-15', '2023-02-05', '2023-01-25', '2023-03-10']}
df_date1 = pd.DataFrame(date_data1)
df_date1['Date'] = pd.to_datetime(date_data1['Date'])
```

```python
def dateRange(df,startDate,endDate):
    newDate = (df['Date']>startDate) & (df['Date']<endDate)
    df_filter = df[newDate]
    return df_filter

startDate = '2023-01-01' 
endDate = '2023-01-31'
filter_df = dateRange(df_date1,startDate,endDate)
print(filter_df)
```
