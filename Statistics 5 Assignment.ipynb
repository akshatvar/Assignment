```python
#Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

#Solution
"""
Estimation in statistics involves making educated guesses or approximations about population parameters
(such as mean, variance, proportion) based on sample data. Estimation is crucial when it's impractical or
impossible to gather data from an entire population.

Point Estimate: A point estimate is a single value that's used to approximate the value of a population parameter.
It's often calculated from sample data and provides a "best guess" of the true population parameter. For example,
the sample mean is a point estimate of the population mean.

Interval Estimate (Confidence Interval): An interval estimate provides a range of values within which the true population
parameter is likely to fall, along with a level of confidence. It accounts for the uncertainty associated with estimation.
A common form of interval estimate is the confidence interval, which provides a range of values around the point estimate.
For instance, a 95% confidence interval for the population mean indicates that we're 95% confident that the true population
mean lies within that interval.
"""
```

```python
#Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

#Solution
"""
Hypothesis testing is a statistical method used to make decisions about population parameters based on sample data. 
It involves formulating two competing hypotheses, the null hypothesis (H0) and the alternative hypothesis (Ha), and 
then using sample data to determine whether there's enough evidence to reject the null hypothesis in favor of the
alternative hypothesis.

Hypothesis testing is used to:

Make informed decisions about population parameters when only sample data is available.
Assess the validity of assumptions or claims made about a population.
Provide a structured framework for drawing conclusions from data.
Enhance the credibility and reliability of research findings.
"""
```

```python
#Q5: Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.

#Solution
import scipy.stats as stats

sample1 = [32,52,90,12,10,45,29,39]
sample2 = [43,88,93,50,29,27,56,72]

t_statistic, p_value = stats.ttest_ind(sample1,sample2)

alpha = 0.05 #significnce level

if p_value<alpha:
    print("Null Hypothesis rejected")

else:
    print("Null Hypothesis accepted")
```

```python
#Q6: What is a null and alternative hypothesis? Give some examples.

#Solution
"""
Null Hypothesis (H0): The null hypothesis is a statement that there is no significant difference or effect. It's the default
assumption that's tested against the alternative hypothesis. For example, "The average height of male and female students is the same."

Alternative Hypothesis (Ha): The alternative hypothesis is a statement that contradicts the null hypothesis and suggests a
significant difference or effect. For example, "The average height of male students is greater than the average height of female students."
"""
```

```python
#Q7: Write down the steps involved in hypothesis testing.

#Solution
"""
Formulate the null hypothesis (H0) and the alternative hypothesis (Ha).
Choose an appropriate significance level (alpha) to determine the threshold for rejecting the null hypothesis.
Collect sample data relevant to the hypothesis being tested.
Calculate the appropriate test statistic based on the type of test (e.g., t-test, z-test, chi-squared test).
Determine the p-value associated with the test statistic.
Compare the p-value to the significance level (alpha):
If p-value < alpha, reject the null hypothesis.
If p-value >= alpha, fail to reject the null hypothesis.
Draw a conclusion based on the comparison:
If the null hypothesis is rejected, there's evidence to support the alternative hypothesis.
If the null hypothesis is not rejected, the data does not provide sufficient evidence to support the alternative hypothesis.
"""
```

```python
#Q8: Define p-value and explain its significance in hypothesis testing.

#Solution
"""
The p-value, or probability value, is a measure of the strength of evidence against the null hypothesis in hypothesis testing.
It represents the probability of observing the test statistic (or a more extreme value) under the assumption that the null hypothesis
is true.

The significance of the p-value lies in its ability to quantify the strength of evidence against the null hypothesis. A common threshold
for significance is the alpha level (often set at 0.05). If the p-value is less than or equal to the alpha level, it suggests that the
observed data provides enough evidence to reject the null hypothesis. However, the final decision should also consider the context,
potential biases, and practical implications of the results.
"""
```

```python
#Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom 
#parameter set to 10.

#Solution
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

dof = 10 #degree of freedom
x = np.linspace(-6,6,600)
y = stats.t.pdf(x,dof)
plt.plot(x,y,label = f"t-distribution (dof = {dof})")
plt.title("Student's t-Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
# Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two 
# random samples of equal size and a null hypothesis that the population means are equal.

# Solution
import numpy as np
import scipy.stats as stats

sample1 = np.random.normal(loc=89,scale=12,size=40)
sample2 = np.random.normal(loc=83,scale=10,size=40)

t_statistic,p_value = stats.ttest_ind(sample1,sample2)

alpha = 0.05
print("Two-Sample T-Test Results")
print("Sample 1 mean:", np.mean(sample1))
print("Sample 2 mean:", np.mean(sample2))
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

if p_value<alpha:
    print("Null Hypothesis rejected")

else:
    print("Null Hypothesis accepted")
```

```python
Q11: What is Student’s t distribution? When to use the t-Distribution.

#Solution
"""
The Student's t-distribution (often simply referred to as the t-distribution) is a probability distribution that is used in
statistical hypothesis testing and confidence interval estimation. It's similar in shape to the normal distribution but has
heavier tails, which means that it accounts for greater variability in smaller sample sizes.

The t-distribution arises when dealing with small sample sizes and unknown population standard deviations. It's used when the
assumptions of the normal distribution (specifically, a known population standard deviation) are not met, and the population
variance needs to be estimated from the sample data.

The t-distribution is especially valuable in cases where sample sizes are small (typically n < 30) or when the population standard
deviation is unknown. It becomes more similar to the standard normal distribution (z-distribution) as the sample size increases.
"""
```
