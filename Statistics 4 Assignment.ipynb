```python
#Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

#Solution
"""
Probability Mass Function (PMF) and Probability Density Function (PDF) are concepts used in probability theory and statistics
to describe the distribution of random variables.

Probability Mass Function (PMF): This is applicable to discrete random variables. It gives the probability that a discrete random
variable takes on a specific value. Mathematically, for a random variable X and a specific value x, PMF is denoted as P(X = x).

Probability Density Function (PDF): This is applicable to continuous random variables. Unlike PMF, which gives the probability at
a specific point, PDF gives the probability that a continuous random variable falls within a certain range. The integral of the PDF
over a range gives the probability of the variable falling within that range.

Example:
Let's consider a fair six-sided die. The random variable X represents the outcome of rolling the die. The PMF for this die would be:
P(X = x) = 1/6 for x = 1, 2, 3, 4, 5, 6 (because all outcomes are equally likely).
The PDF for this continuous variable would describe the likelihood of finding an individual with a specific height. However,
since height can take on a wide range of values, we can talk about the probability of finding someone with a height between,
say, 160 cm and 170 cm, by integrating the PDF over that range.
"""
```

```python
#Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

#Solutiion
"""
Cumulative Density Function (CDF) is a function that gives the probability that a random variable takes on a value less than
or equal to a specific value. It provides a way to understand the entire distribution of a random variable, including both
discrete and continuous cases.

Example:
Consider the same fair six-sided die. The CDF for this die would be:

For x ≤ 1: CDF(X ≤ x) = 1/6
For x ≤ 2: CDF(X ≤ x) = 2/6

For x ≤ 6: CDF(X ≤ x) = 1
The CDF provides a clearer picture of how the probabilities accumulate as you move along the values of the random variable.

CDF is used because it's a useful tool to calculate probabilities for a range of values, and it's often easier to work with
than directly using the PMF or PDF. It helps in various statistical calculations and hypothesis testing.

"""
```

```python
#Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters
#of the normal distribution relate to the shape of the distribution.

#Solution
"""
The normal distribution, also known as the Gaussian distribution, is a fundamental probability distribution that is used to model
a wide variety of real-world phenomena. Some examples of situations where the normal distribution might be used as a model include:

Height of Individuals: Human height tends to follow a roughly normal distribution, with most people clustered around the average height.

Measurement Errors: Many measurement errors and natural variations in processes also tend to follow a normal distribution.

Test Scores: In standardized testing, scores often follow a normal distribution, with most students scoring around the average.

Financial Returns: Stock market returns and other financial data often exhibit approximately normal distribution-like behavior.

The normal distribution is defined by two parameters: the mean (μ) and the standard deviation (σ). The mean represents the central location
of the distribution, and the standard deviation controls the spread or dispersion of the distribution.

A larger standard deviation results in a wider distribution.
A smaller standard deviation results in a narrower distribution.
The shape of the normal distribution is symmetric around the mean, and it's a bell-shaped curve. The probability that a random variable
falls within a certain range is determined by the area under the curve within that range.
"""
```

```python
#Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal Distribution.

#Solution
"""
The normal distribution is of paramount importance in statistics and probability theory due to its remarkable properties and
wide applicability. Some reasons for its importance include:

Central Limit Theorem: The sums or averages of a large number of independent, identically distributed random variables tend to
follow a normal distribution, regardless of the original distribution. This theorem underpins many statistical analyses.

Inferential Statistics: Many statistical methods and tests assume that the data follows a normal distribution. This simplifies calculations
and helps in making accurate inferences about populations.

Data Modeling: The normal distribution provides a good approximation for a wide range of real-world phenomena, making it a useful tool 
for modeling and prediction.

Standardized Scores (Z-scores): Z-scores are a way to standardize data based on the mean and standard deviation of the normal distribution. 
They allow for easy comparison and analysis of data points.

Examples of real-life situations exhibiting normal distribution include:

IQ scores: IQ scores tend to follow a normal distribution, with most people clustered around the average IQ of 100.
Heights of individuals: In a population, heights generally follow a normal distribution, with most people around the average height.
Errors in measurement: Measurement errors often exhibit a normal distribution pattern, which is useful for error analysis.
"""
```

```python
#Q5: What is Bernoulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and
#Binomial Distribution?

#Solution
"""
Bernoulli Distribution:
The Bernoulli distribution is a discrete probability distribution that models a random experiment with only two possible
outcomes: success (typically denoted as 1) or failure (typically denoted as 0). It's used for situations where an event
can have one of two outcomes with a fixed probability of success.

Example:
Consider flipping a coin. Let's say we define "heads" as success (1) and "tails" as failure (0). The outcome of the coin flip
follows a Bernoulli distribution.

Difference between Bernoulli and Binomial Distribution:
The key difference between the Bernoulli distribution and the binomial distribution lies in the number of trials. The binomial
distribution extends the concept of the Bernoulli distribution to situations where a series of independent Bernoulli trials are performed.

Bernoulli Distribution: Models a single trial with two outcomes (success or failure).
Binomial Distribution: Models the number of successes in a fixed number of independent Bernoulli trials.
The binomial distribution has two additional parameters: the number of trials (n) and the probability of success (p). It gives the probability
of obtaining exactly k successes in n trials.
"""
```

```python
#Q6: Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed,
#what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your
#calculations.

#Solution
"""
To calculate this probability, we need to find the z-score for the value 60 and then use the standard normal distribution (z-distribution)
to find the corresponding cumulative probability.

The formula for calculating the z-score is: z = (X - μ) / σ
Where:

X = Value (60 in this case)
μ = Mean (50)
σ = Standard deviation (10)
Calculating the z-score: z = (60 - 50) / 10 = 1

Now, we need to find the cumulative probability for z = 1 using a standard normal distribution table or calculator.
The cumulative probability for z = 1 is approximately 0.8413.

So, the probability that a randomly selected observation will be greater than 60 is approximately 1 - 0.8413 = 0.1587, or 15.87%.
"""
```

```python
#Q7: Explain uniform Distribution with an example.

#Solution
"""
The uniform distribution is a probability distribution where every possible outcome has an equal probability of occurring.
In other words, all values within a specified range are equally likely to occur.

Example:
Consider rolling a fair six-sided die. The outcome of rolling the die follows a uniform distribution because each of the six
faces has an equal probability of landing face up. The probability of rolling any particular number (1, 2, 3, 4, 5, or 6) is
1/6, and this probability is the same for each possible outcome.

In the uniform distribution, the probability density function (PDF) is a constant within the specified range and zero outside
that range. This leads to a flat, constant shape for the distribution.
"""
```

```python
#Q8: What is the z score? State the importance of the z score.

#Solution
"""
The z-score, also known as the standard score, is a measure that quantifies how many standard deviations a data point is away
from the mean of a distribution. It's a way to standardize and compare data points from different distributions.

The formula for calculating the z-score is: z = (X - μ) / σ
Where:

X = Data point
μ = Mean of the distribution
σ = Standard deviation of the distribution
Importance of the z-score:

Standardization: Z-scores allow for the comparison of data points from different distributions. This is especially useful in
situations where the scales of measurement differ.

Outlier Detection: Z-scores help identify outliers in a dataset. Data points with z-scores significantly far from zero might indicate
unusual observations.

Probability Calculation: Z-scores are used to find probabilities associated with specific data values in a normal distribution.
They help in understanding how likely or unlikely an observation is.

Data Analysis: Z-scores are used in hypothesis testing, confidence interval estimation, and various statistical analyses to make
comparisons and draw conclusions about data.
"""
```

```python
#Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

#Solution
"""
Central Limit Theorem (CLT):
The Central Limit Theorem is a fundamental theorem in statistics that states that when independent random variables are added together,
their sum approaches a normal distribution, regardless of the original distribution of the variables. This holds true as long as the
sample size is sufficiently large.

Significance of the Central Limit Theorem:

Approximation of Distributions: The CLT allows us to approximate the distribution of sample means (or sums) to a normal distribution,
even if the individual variables are not normally distributed. This is important because normal distributions are well-understood and
easy to work with.

Statistical Inference: The CLT is the foundation for many statistical techniques, such as hypothesis testing and confidence interval
estimation. It's used when dealing with large sample sizes to make reliable inferences about population parameters.

Real-world Applications: Many natural processes involve the sum of multiple factors. The CLT explains why we often see normal distributions
emerge in real-world scenarios, even if the individual components are not normally distributed.
"""
```

```python
#Q10: State the assumptions of the Central Limit Theorem.

#Solution
"""
The Central Limit Theorem (CLT) is a powerful tool in statistics, but it relies on certain assumptions to hold true. These assumptions include:

Independence: The random variables being summed or averaged must be independent of each other. In other words, the behavior of one
variable should not influence the behavior of another.

Finite Variance: The random variables should have finite variances. This ensures that the sum of their variances doesn't become infinite
as more variables are added.

Sample Size: The sample size should be sufficiently large. While there isn't a strict rule for the minimum sample size, a common guideline
is that the sample size should be at least 30.

No Extreme Skewness: If the distribution of the original variables is very skewed, the CLT might not apply as effectively, especially for
smaller sample sizes.

It's important to note that while these assumptions increase the likelihood that the CLT will hold, the theorem can still provide useful
approximations even if some of these assumptions are not perfectly met.
"""
```
